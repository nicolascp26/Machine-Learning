```python
plt.subplots(figsize=(25, 15))
#Función de correlación de pandas (cercano a 1 es la mejor correlación, negativos la peor)
correlation_matrix = df_bikes.corr().round(2)
sns.heatmap(data=correlation_matrix,annot=True)
```

```python
cols = ['season','month','holiday','weekday','workingday','weather']
for col in cols:
    df_bikes[col]=df_bikes[col].astype('category')
df_bikes.info()
```

```python
correlation_matrix = df_bikes.corr().round(2)
sns.heatmap(data=correlation_matrix,annot=True)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
x = df_bikes.drop(columns=['humidity','windspeed','count'])
y = df_bikes['count']
```

```python
x.info()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=3)
```

```python
modelo.fit(x_train,y_train)
y_test_pred = modelo.predict(x_test)
```

```python
print("----- LINEAR REGRESSION RESULTS -----")
print("Model Accuracy global score: %.3f" % modelo.score(x_test,y_test))
from sklearn.metrics import mean_squared_error
error = np.sqrt(mean_squared_error(y_test,y_test_pred))
print("Percentage of error: %f" % (error*100))
```

```python
# Plot outputs
plt.scatter(x_test['casual'], y_test,  color='black')
plt.plot(x_test['casual'], y_test_pred, color='blue', linewidth=1)
plt.title('Trained Model: Test Prediction')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```
